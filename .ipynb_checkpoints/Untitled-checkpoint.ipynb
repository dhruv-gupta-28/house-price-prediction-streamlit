```python
import pandas as pd

df_train = pd.read_csv("train.csv")
df_test = pd.read_csv("test.csv")

df_train.head()
```

```python
df_train.shape
```

```python
df_train.info()
```

```python
missing = df_train.isnull().sum()
missing = missing[missing > 0].sort_values(ascending=False)

missing.head(10)
```

```python
missing_percent = df_train.isnull().mean() * 100

good_cols = missing_percent[missing_percent < 50].index

df_train = df_train[good_cols]
df_test = df_test[good_cols.drop("SalePrice", errors='ignore')]
```

```python
num_cols = df_train.select_dtypes(include=['int64','float64']).columns
cat_cols = df_train.select_dtypes(include=['object','string']).columns

print("Numerical:", len(num_cols))
print("Categorical:", len(cat_cols))
```

```python
for col in num_cols:
    if col != "SalePrice":
        df_train[col] = df_train[col].fillna(df_train[col].median())
        df_test[col] = df_test[col].fillna(df_train[col].median())
```

```python
df_train[num_cols].isnull().sum().sum()
```

```python
for col in cat_cols:
    df_train[col] = df_train[col].fillna("Missing")
    df_test[col] = df_test[col].fillna("Missing")
```

```python
df_train[cat_cols].isnull().sum().sum()
```

```python
df_train = pd.get_dummies(df_train)
df_test = pd.get_dummies(df_test)
```

```python
df_train, df_test = df_train.align(df_test, join='left', axis=1, fill_value=0)
```

```python
df_train.shape
```

```python
df_test.shape
```

```python
X = df_train.drop("SalePrice", axis=1)
y = df_train["SalePrice"]
```

```python
print("SalePrice in X:", "SalePrice" in X.columns)
print("SalePrice in df_test:", "SalePrice" in df_test.columns)
```

```python
# ---------- STEP 1: OUTLIER REMOVAL ----------
df_train_clean = df_train[df_train["GrLivArea"] < 4000].copy()

# ---------- STEP 2: FEATURE ENGINEERING (BEFORE EVERYTHING) ----------
df_train_clean["TotalSF"] = df_train_clean["TotalBsmtSF"] + df_train_clean["1stFlrSF"] + df_train_clean["2ndFlrSF"]
df_test["TotalSF"] = df_test["TotalBsmtSF"] + df_test["1stFlrSF"] + df_test["2ndFlrSF"]

df_train_clean["TotalBathrooms"] = df_train_clean["FullBath"] + (0.5 * df_train_clean["HalfBath"])
df_test["TotalBathrooms"] = df_test["FullBath"] + (0.5 * df_test["HalfBath"])
```

```python
df_test = df_test.drop("SalePrice", axis=1, errors='ignore')
```

```python
print("SalePrice in df_test:", "SalePrice" in df_test.columns)
```

```python
corr = df_train.corr()["SalePrice"].abs().sort_values(ascending=False)

corr.head(15)
```

```python
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Remove outliers (important)
df_train_clean = df_train[df_train["GrLivArea"] < 4000]

# Feature engineering
df_train_clean["TotalSF"] = df_train_clean["TotalBsmtSF"] + df_train_clean["1stFlrSF"] + df_train_clean["2ndFlrSF"]
df_test["TotalSF"] = df_test["TotalBsmtSF"] + df_test["1stFlrSF"] + df_test["2ndFlrSF"]

df_train_clean["TotalBathrooms"] = df_train_clean["FullBath"] + (0.5 * df_train_clean["HalfBath"])
df_test["TotalBathrooms"] = df_test["FullBath"] + (0.5 * df_test["HalfBath"])

# Separate
X = df_train_clean.drop("SalePrice", axis=1)
y = df_train_clean["SalePrice"]

# Scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
df_test_scaled = scaler.transform(df_test)

# Split
X_train, X_val, y_train, y_val = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)

# Log transform
y_train = np.log(y_train)
y_val = np.log(y_val)

# Train
model = LinearRegression()
model.fit(X_train, y_train)

# Predict
y_pred = model.predict(X_val)

# Convert back
y_pred = np.exp(y_pred)
y_val_actual = np.exp(y_val)

# RMSE
rmse = np.sqrt(mean_squared_error(y_val_actual, y_pred))
print("RMSE:", rmse)
```

```python
# Predict
predictions = model.predict(df_test_scaled)

# Convert back
predictions = np.exp(predictions)
```

```python
submission = pd.DataFrame({
    "Id": df_test["Id"],
    "SalePrice": predictions
})

submission.to_csv("submission.csv", index=False)
```
